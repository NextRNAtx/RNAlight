- Note: 
        - RNAs longer than the fixed length were truncated to the fixed length and replaced by the part with original 5 prime sequence. 
        - RNAs shorter than the fixed length were padding with 'N' before 5 prime or after 3 prime of RNA sequence
        

        AUROCs of padding N before 5 prime  are from lncRNA/5_prime_padding/01_DL_Model_Output of each model
        AUROCs of padding N after 3 prime  are from   RNAlight_Private/lncRNA/03_Model_Construction/02_Deep_Learning_Model/01_DL_Model_Output of each model

- Cross valiadation: highest mean_auroc_on_CV in CV_result.tsv
-  Test: AUROC in Evaluate_Result_TestSet.txt

```r
lncRNA_CV_df <- rbind(lncRNA_CNN_5_prime_N_CV,lncRNA_CNN_RNN_5_prime_N_CV,lncRNA_CNN_3_prime_N_CV,lncRNA_CNN_RNN_3_prime_N_CV)
lncRNA_CV_df$model <- factor(lncRNA_CV_df$model,levels=c("CNN","CNN_RNN"))
lncRNA_CV_df$padding <- factor(lncRNA_CV_df$padding,levels=c("Three","Five"))
```
