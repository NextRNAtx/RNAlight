## 1. Whole genome lncRNA, mRNA, snoRN and snRNA

```r
# Load file
Whole_genome_lncRNA <- read.csv("./lncRNA_whole_genome/Whole_genome_lncRNA_predict_df.tsv",sep = '\t')
Whole_genome_mRNA <- read.csv("./mRNA_whole_genome/Whole_genome_mRNA_predict_df.tsv",sep = '\t')
Whole_genome_snoRNA <- read.csv("./snoRNA_whole_genome/snoRNAs_predict_df.tsv",sep = '\t')
Whole_genome_snRNA <- read.csv("./snRNA_whole_genome/snRNAs_predict_df.tsv",sep = '\t')
```

```r
# Statistic numbers
colnames(Whole_genome_lncRNA)

nrow(Whole_genome_lncRNA)
nrow(Whole_genome_mRNA)
nrow(Whole_genome_snoRNA)
nrow(Whole_genome_snRNA)
```

### 1.1 lncRNA

```r
# Scale RNALight predicted probability as Light score
Whole_genome_lncRNA_df <- as.data.frame(Whole_genome_lncRNA$RNALight_pred_prob) *2 -1
colnames(Whole_genome_lncRNA_df) <- "Light_score"
Whole_genome_lncRNA_df$group <- "lncRNA"

# Density plot with text
p_density <- ggplot(Whole_genome_lncRNA_df ,aes(x=Light_score,fill = group))+geom_density(alpha = 0.6) + 
theme_bw() +
theme(axis.text = element_text(size=14),
      axis.title = element_text(size=14),
     panel.grid = element_blank(),
     panel.border = element_rect(fill = NA,color = "black",size =2,linetype = "solid"))+
xlim(-1,1)+
ylim(0,1)+
scale_fill_manual(values=c("red3"))+
coord_fixed(ratio =  0.8)
ggsave("./01_Light_score_distribution_Output/01_RNALight_lncRNA_with_text.pdf", p_density, width = 10, height = 10)

p_density

# Density plot 
p_density <- p_density + theme(axis.text = element_blank(),axis.title = element_blank(),axis.ticks = element_blank(),legend.position = "none")
ggsave("./01_Light_score_distribution_Output/01_RNALight_lncRNA.pdf", p_density, width = 10, height = 10) 

p_density
```

### 1.2 mRNA

```r
# Scale RNALight predicted probability as Light score
Whole_genome_mRNA_df <- as.data.frame(Whole_genome_mRNA$RNALight_pred_prob) *2 -1
colnames(Whole_genome_mRNA_df) <- "Light_score"
Whole_genome_mRNA_df$group <- "mRNA"

# Density plot with text
p_density <- ggplot(Whole_genome_mRNA_df ,aes(x=Light_score,fill = group))+geom_density(alpha = 0.6) + 
theme_bw() +
theme(axis.text = element_text(size=14),
      axis.title = element_text(size=14),
     panel.grid = element_blank(),
     panel.border = element_rect(fill = NA,color = "black",size =2,linetype = "solid"))+
xlim(-1,1)+
ylim(0,1)+
scale_fill_manual(values=c("royalblue3"))+
coord_fixed(ratio =  0.8)
ggsave("./01_Light_score_distribution_Output/01_RNALight_mRNA_with_text.pdf", p_density, width = 10, height = 10)

p_density

# Density plot 
p_density <- p_density + theme(axis.text = element_blank(),axis.title = element_blank(),axis.ticks = element_blank(),legend.position = "none")
ggsave("./01_Light_score_distribution_Output/01_RNALight_mRNA.pdf", p_density, width = 10, height = 10) 

p_density
```

### 1.5 Combine as Boxplot

```r
# Combine files
Whole_genome_RNA_Light_score_df_long<- rbind(Whole_genome_lncRNA_df,Whole_genome_mRNA_df,Whole_genome_snoRNA_df,Whole_genome_snRNA_df)
Whole_genome_RNA_Light_score_df_long$group <- factor(Whole_genome_RNA_Light_score_df_long$group,levels = c("snRNA","snoRNA","mRNA","lncRNA"))

# Light score boxplot with text
p_box <- ggplot(data=Whole_genome_RNA_Light_score_df_long,aes(x=group,y=Light_score,fill = group))+
geom_boxplot(width = 0.3,outlier.color="white",alpha = 0.6,position=position_dodge(width=0.5))+
theme_classic()+
scale_fill_manual(values = c("#ff9770","#ff7096","royalblue3","red3"))+
theme_classic()+
theme(axis.text = element_text(size=14),
      axis.title = element_text(size=14),
      element_line(size = 1.2))+
coord_flip()
ggsave("./01_Light_score_distribution_Output/01_Whole_genome_RNA_Light_score_boxplot_with_text.pdf", p_box, width = 10, height = 12) 

p_box

# Light score boxplot
p_box2 <- p_box + theme(axis.text = element_blank(),axis.title = element_blank(),axis.ticks = element_blank(),legend.position = "None")
ggsave("./01_Light_score_distribution_Output/01_Whole_genome_RNA_Light_score_boxplot.pdf", p_box2, width = 10, height = 10) 

p_box2
```
