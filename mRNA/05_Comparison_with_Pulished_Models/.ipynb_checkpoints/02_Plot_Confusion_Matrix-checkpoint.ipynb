```r
library(ggplot2)
```

# LncLight

```r
LncLight_test_set <- read.csv("./LncLight/Evaluation_Result/lncRNA_sublocation_TestSet_LncLight_predict.tsv",sep = '\t')
LncLight_test_set_df <- LncLight_test_set[,c(1,4,5)]
LncLight_test_set_df$tag <- factor(LncLight_test_set_df$tag,levels = c(1,0),labels= c("Nuc","Cyto"))
LncLight_test_set_df$predict_label <- factor(LncLight_test_set_df$predict_label,levels = c(1,0),labels= c("Nuc","Cyto"))
LncLight_freq = as.data.frame(table(LncLight_test_set_df[,2:3]))
```

```r
LncLight_freq
```

```r
p <- ggplot(data =LncLight_freq,aes(x=predict_label,y=factor(tag,levels = c("Cyto","Nuc")),fill=Freq)) + 
geom_tile()+
theme_classic()+
theme(panel.border = element_rect(colour = "black", fill=NA,size = 1.5))+
scale_y_discrete(expand = c(0,0))+
scale_x_discrete(expand = c(0,0))+
scale_fill_gradient(low = "white", high = "#B71C1C")
ggsave("./Confusion_matrix/LncLight_with_text.pdf", p, width = 10, height = 10) 
```

```r
p
```

```r
p <- p +theme(legend.key.height = unit(5,"cm"),
         legend.key.width = unit(2,"cm"),
         legend.text = element_text(size = 0),
         legend.title = element_text(size = 0),
         axis.text = element_blank(),
         axis.title = element_blank(),
         axis.ticks = element_blank())
ggsave("./Confusion_matrix/LncLight_with_bar.pdf", p, width = 10, height = 10) 
```

```r
p
```

```r
p <- p +theme(legend.position = "None",
         axis.text = element_blank(),
         axis.title = element_blank(),
         axis.ticks = element_blank())
ggsave("./Confusion_matrix/LncLight.pdf", p, width = 10, height = 10) 
```

```r
p 
```
