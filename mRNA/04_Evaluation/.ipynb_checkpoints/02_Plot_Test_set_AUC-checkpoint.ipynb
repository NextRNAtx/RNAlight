```r
library(ggplot2)
```

```r
LncLight_lgb <- read.csv('./20210208_cv_test/LightGBM/AUROC_info.txt',sep = '\t')
LncLight_SVM <- read.csv('./20210208_cv_test/SVM/AUROC_info.txt',sep = '\t')
LncLight_lr <- read.csv('./20210208_cv_test/LogisticRegression/AUROC_info.txt',sep = '\t')
LncLight_CNN <- read.csv('./CNN_4000nt/Code/CNN_4000nt_Model_Output/AUROC_info.txt',sep = '\t')
LncLight_CNN_RNN <- read.csv('./CNN_RNN_4000nt/Code/CNN_RNN_4000nt_Model_Output/AUROC_info.txt',sep = '\t')
LncLight_RNN <- read.csv('./RNN_4000nt/Code/RNN_4000nt_Model_Output/AUROC_info.txt',sep = '\t')


LncLight_lgb$group <- "lgb"
LncLight_SVM$group <- "SVM"
LncLight_lr $group <- "lr"
LncLight_CNN$group <- "CNN"
LncLight_CNN_RNN$group <- "CNN_RNN"
LncLight_RNN$group <- "RNN"
```

```r
LncLight_dataset <- rbind(LncLight_lgb,LncLight_SVM,LncLight_lr,LncLight_CNN,LncLight_CNN_RNN,LncLight_RNN)
```

```r
LncLight_dataset$group<- factor(LncLight_dataset$group,
                levels = c("lgb",  "SVM","lr","CNN","CNN_RNN","RNN"))
```

```r
p_auc_LncLight <- ggplot(data=LncLight_dataset ,aes(x=FPR,y=TPR))+
geom_line(aes(color=group),size = 1.2)+
geom_segment(x=0,xend=1,y=0,yend=1,colour ="gray",linetype = "dashed")+
theme_classic()+
theme(panel.border = element_rect(colour = "black", fill=NA,size = 2),
      axis.text = element_text(size=14),
      axis.title = element_text(size=14))+
scale_color_manual(values = c("#E22146","#F9DB78","#DAC2D1","#CADEB8","#A5C2E3","#EAB28A"))
ggsave("./Model_AUROC/LncLight_auc_without_RF_with_text.pdf", p_auc_LncLight, width = 10, height = 10) 
```

```r
p_auc_LncLight
```

```r
p_auc_LncLight <- p_auc_LncLight + theme(axis.text = element_blank(),axis.title = element_blank(),axis.ticks = element_blank(),legend.position = "None")
ggsave("./Model_AUROC/LncLight_auc_without_RF.pdf", p_auc_LncLight, width = 10, height = 10) 
```

```r
LncLight_auc_df <- data.frame(list("auc" = c(0.78330190771961,0.754547471162378,0.756127994676131,0.712095164152618,0.70915594498669,0.630129769299024),
                                  "model" = c("LightGBM","SVM","LR","CNN","CNN_RNN","RNN")))
```

```r
LncLight_auc_df$model<- factor(LncLight_auc_df$model,
                levels = c("LightGBM","SVM","LR","CNN","CNN_RNN","RNN"))
```

```r
p_auc_LncLight_bar <- ggplot(LncLight_auc_df ,aes(x=model,y=auc,fill = model))+
geom_bar(stat="identity",position=position_dodge(0.7),width=0.6) +
ylim(0,1) +
theme_classic()+
theme(axis.text = element_text(size=14),
      axis.title = element_text(size=14),
      element_line(size = 1.2))+
scale_fill_manual(values = c("#E22146","#F9DB78","#DAC2D1","#CADEB8","#A5C2E3","#EAB28A"))
ggsave("./Model_AUROC/LncLight_auc_bar_without_with_text.pdf",p_auc_LncLight_bar, width = 10, height = 10) 
```
