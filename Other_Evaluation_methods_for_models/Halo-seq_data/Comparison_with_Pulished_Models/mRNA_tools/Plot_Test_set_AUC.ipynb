```r
library(ggplot2)
```

```r
# Load AUROC file
RNALight <- read.csv('./RNA-Light_mRNA/Evaluation_Result/AUROC_info.txt',sep = '\t')
iLoc_mRNA <- read.csv('./iLoc-mRNA/Linux/Evaluation_Result/AUROC_info.txt',sep = '\t')

# Group
RNALight$group <- "RNAlight"
iLoc_mRNA$group <- "iLoc_mRNA"



AUROC_df <- rbind(RNALight,iLoc_mRNA)
AUROC_df$group<- factor(AUROC_df$group,
                levels = c("RNAlight",  "iLoc_mRNA"))
```

```r
# Plot
p_auc <- ggplot(data=AUROC_df ,aes(x=FPR,y=TPR))+
geom_line(aes(color=group),size = 4)+
geom_segment(x=0,xend=1,y=0,yend=1,colour ="gray",linetype = "dashed",size = 4)+
theme_classic()+
theme(panel.border = element_rect(colour = "black", fill=NA,size = 2),
      axis.text = element_text(size=14),
      axis.title = element_text(size=14))+
scale_color_manual(values = c("#0176B5","#50B2D4"))
ggsave("./03_RNA-Light_Halo_seq_AUROC_with_text.pdf", p_auc, width = 10, height = 10) 
p_auc
```

```r
p_auc<- p_auc + theme(axis.text = element_blank(),axis.title = element_blank(),axis.ticks = element_blank(),legend.position = "None")
ggsave("./03_RNA-Light_Halo_seq_AUROC.pdf", p_auc, width = 10, height = 10) 
p_auc
```

```r
# Load AUPRC file
RNALight <- read.csv('./RNA-Light_mRNA/Evaluation_Result/AUPRC_info.txt',sep = '\t')
iLoc_mRNA <- read.csv('./iLoc-mRNA/Linux/Evaluation_Result/AUPRC_info.txt',sep = '\t')

# Group
RNALight$group <- "RNAlight"
iLoc_mRNA$group <- "iLoc_mRNA"



AUPRC_df <- rbind(RNALight,iLoc_mRNA)
AUPRC_df$group<- factor(AUROC_df$group,
                levels = c("RNAlight",  "iLoc_mRNA"))
```

```r
# Plot
p_auprc <- ggplot(data=AUPRC_df ,aes(x=Recall,y=Precision))+
geom_line(aes(color=group),size = 4)+
ylim(0,1)+
theme_classic()+
theme(panel.border = element_rect(colour = "black", fill=NA,size = 2),
      axis.text = element_text(size=14),
      axis.title = element_text(size=14))+
scale_color_manual(values = c("#0176B5","#50B2D4"))
ggsave("./03_RNA-Light_Halo_seq_AUPRC_with_text.pdf", p_auprc, width = 10, height = 10) 
p_auprc
```

```r
p_auprc<- p_auprc + theme(axis.text = element_blank(),axis.title = element_blank(),axis.ticks = element_blank(),legend.position = "None")
ggsave("./03_RNA-Light_Halo_seq_AUPRC.pdf", p_auprc, width = 10, height = 10) 
p_auprc
```
